# 1. PRICE DISTRIBUTION

```python
sns.distplot(ac['Price'])
font = {'family':'serif','color':'red','size':20}
plt.title('Price Distribution',fontdict=font)
plt.show()
```

# 2. BRAND POPULARITY

```python
plt.figure(figsize=(20,5))
sns.countplot(data=ac, x='Brand',palette='rainbow')
font = {'family':'serif','color':'green','size':30}
plt.title('Brand Recognition',fontdict=font)
plt.xticks(rotation=90)
plt.show()
```

# 3. Which Types of Collecation In Cars

```python
plt.figure(figsize=(20,5))
sns.countplot(data=ac, x = 'Fuel',)
font = {'family':'serif','color':'black','size':25}
plt.title('Fuel Type Distribution',fontdict=font)
plt.show()

```

# 4. Which Types of Car In Australia

```python
plt.figure(figsize=(20,5))
sns.countplot(x='Type',data=ac,color='olive')
plt.xticks(rotation=90)
plt.show()
```

# 5. Check Car Status

```python
sns.countplot(x = 'Status',data=ac,palette='rainbow')
plt.show()
```

# 6. Variations In CC's Of Cars

```python
plt.figure(figsize=(10,5))
sns.histplot(ac['CC'])
plt.show()
```

# 7. Seating Capacity Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(ac['Seating Capacity'])
plt.show()
```

# 8. Price Distribution With Respect To Each Brand

```python
plt.figure(figsize=(20,6))
sns.boxplot(x='Brand', y='Price',data=ac,palette='spring')
plt.xticks(rotation=90)
plt.show()
```

# 9. CC Engine Distribution With Respect TO Each Brand

```python
plt.figure(figsize=(20,5))
sns.boxplot(x='Brand', y='CC',data=ac,palette='winter')
plt.xticks(rotation=90)
plt.show()
```
